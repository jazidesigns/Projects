### Import Libraries

```python
import pandas as pd
```

### Import Dataset

```python
df = pd.read_csv("user_reviews.csv")
df.sample(5)
```

### Add Column len_text

```python
df["len_text"] = df["text"].str.len()
df.sample(5)
```

### New column based on multiple conditions and existing column values

**a) For Loop**

```python
super_category = []

for x in df[["grade", "len_text"]].values:
    if x[0] >= 9 and x[1] >= 1000:
        super_category.append("expert reviewer")
        
    elif x[0] <= 1 and x[1] >= 1000:
        super_category.append("opposed-reviewer")
        
    else:
        super_category.append("neutral reviewer")

df["super category"] = super_category
df.sample(5)
```

**b) While Loop**

**c) Vectorisation** 

```python
def review(vec):
    grade = vec[0]
    len_text = vec[1]
    
    if grade >= 9 and len_text >= 1000:
        return "expert reviewer"
        
    elif grade <= 1 and len_text >= 1000:
        return "opposed-reviewer"
        
    else:
        return "neutral reviewer"
    
df["super category"] = df[["grade", "len_text"]].apply(review, axis=1)
df.sample(5)
```
